```python
# Exercice 1
import numpy as np

# -----------------------------------------------------------------
# 1. Tableau contenant les nombres de 0 à 20
print("1 --------------------------")
tableau1 = np.arange(21)
print(tableau1)
print(tableau1.shape)

# -----------------------------------------------------------------
# 2. Tableau 3×3 rempli de 1
print("2 --------------------------")
tableau2 = np.ones((3, 3))
print(tableau2)
print(tableau2.shape)

# -----------------------------------------------------------------
# 3. Tableau 4×4 de nombres aléatoires entre 0 et 1
print("3 --------------------------")
tableau3 = np.random.rand(4, 4)
print(np.round(tableau3, 2))
print(tableau3.shape)

# -----------------------------------------------------------------
# 4. Sous-matrice centrale 2x2
print("4 --------------------------")
sous_matrice = tableau3[1:3, 1:3]  # Extrait les lignes 1 à 2 et colonnes 1 à 2
print(np.round(sous_matrice, 2))
print(sous_matrice.shape)

# -----------------------------------------------------------------
# Exercice 2 : Opérations et statistiques
print("Exercice 2 --------------------------")
data = np.array([12, 5, 8, 20, 15, 3, 10, 18])

# Calcul de la moyenne des données
moyenne = np.mean(data)
print("La moyenne :", moyenne)

# Calcul de la médiane
mediane = np.median(data)
print("La médiane :", mediane)

# Calcul de l'écart-type
ecart_type = np.std(data)
print("L'écart-type :", ecart_type)

# Recherche de la valeur maximale
max_value = np.max(data)
print("La valeur max :", max_value)

# Recherche de la position (index) de la valeur maximale
max_position = np.argmax(data)
print("La position du max :", max_position)

# Normalisation
data_normalise = (data - np.min(data)) / (np.max(data) - np.min(data))
print("Données normalisées :", data_normalise)

# -----------------------------------------------------------------
# Exercice 3 : Manipulation avancée
print("Exercice 3 --------------------------")

# 1. Générer un tableau de 100 éléments entre 0 et 10 (inclus)
tab = np.linspace(0, 10, 100)
print("Tableau 100 éléments entre 0 et 10 (inclus):", np.round(tab, 2))

# 2a. Extraire les valeurs strictement supérieures à 5
val_sup5 = tab[tab > 5]
print("Les valeurs strictement supérieures à 5:", np.round(val_sup5, 2))

# 2b. Extraire les valeurs paires
entiers = np.arange(100)
paires = entiers[entiers % 2 == 0]
print("Les valeurs paires:", paires)

# 3. Calculer la somme des valeurs impaires
somme_impaires = sum(entiers[entiers % 2 != 0])
print("La somme des valeurs impaires:", somme_impaires)

# -----------------------------------------------------------------
# Partie 2 : Pandas
print("Partie 2 : Pandas --------------------------")
import pandas as pd

data = {
    "Nom": ["Ali", "Sara", "Mohamed", "Aya", "Hiba", "Kamal"],
    "Age": [20, 22, 19, 21, 23, 20],
    "Note": [14, 16, 12, 18, 15, 10],
    "Ville": ["Oujda", "Rabat", "Casablanca", "Fès", "Oujda", "Agadir"]
}

df = pd.DataFrame(data)
print("DataFrame complet:")
print(df)

# -----------------------------------------------------------------
print("Exercice 3 (Pandas) --------------------------")

# Afficher les 3 premières lignes du DataFrame
print("Les 3 premières lignes:")
print(df.head(3))
print(" -------------------------- \n ")
# Afficher le résumé statistique des colonnes numériques
print("Le résumé statistique:")
print(df.describe())
print(" -------------------------- \n ")
# Afficher les types de données de chaque colonne
print("Les types des colonnes:")
print(df.dtypes)
print(" -------------------------- \n ")

# Compter le nombre d'occurrences de chaque ville
print("Répartition par ville:")
print(df["Ville"].value_counts())
print(" -------------------------- \n ")

# Trouver l'étudiant avec la note minimale
print("Étudiant avec la note minimale:")
print(df.loc[df['Note'].idxmin()])

# -----------------------------------------------------------------
# Exercice 5 : Filtrage de données
print("=== EXERCICE 5 : FILTRAGE DE DONNÉES ===")

# 1. Filtrer les étudiants ayant une note >= 15
df1 = df[df["Note"] >= 15]
print("1. Étudiants avec note >= 15 :")
print(df1, '\n')

# 2. Filtrer les étudiants originaires d'Oujda
df2 = df[df["Ville"] == "Oujda"]
print("2. Étudiants originaires d'Oujda :")
print(df2, '\n')

# 3. Filtrer les étudiants de plus de 20 ans ET avec note >= 14
df3 = df[(df['Age'] > 20) & (df['Note'] >= 14)]
print("3. Étudiants > 20 ans ET note >= 14 :")
print(df3, '\n')

# -----------------------------------------------------------------
# Exercice 6 : Transformation et tri des données
print("=== EXERCICE 6 : TRANSFORMATION ET TRI ===")

# 1. Ajouter une colonne "Moyenne_finale" (Note + 2 points bonus)
df["Moyenne_finale"] = df['Note'] + 2
print("1. DataFrame avec Moyenne_finale ajoutée :")
print(df, '\n')

# 2. Filtrer pour garder seulement les étudiants avec Note >= 12
df_clean = df[df['Note'] >= 12]
print("2. DataFrame nettoyé (Note >= 12) :")
print(df_clean, '\n')

# 3. Trier le DataFrame nettoyé par âge décroissant
sd_sorted = df_clean.sort_values(by="Age", ascending=False)
print("3. DataFrame trié par âge décroissant :")
print(sd_sorted, '\n')

# -----------------------------------------------------------------
# Exercice 7 : Visualisation avec Matplotlib
print("=== EXERCICE 7 : VISUALISATION ===")
import matplotlib.pyplot as plt
print(" 1. Histogramme des notes -------------------------- \n ")

# 1. Histogramme des notes
plt.hist(df["Note"])
plt.title("Répartition des notes")
plt.xlabel("Note")
plt.ylabel("Effectif")
plt.show()
print("  2. Diagramme en barres : Âge par étudiant -------------------------- \n ")

# 2. Diagramme en barres : Âge par étudiant
plt.bar(df["Nom"], df['Age'])  # Barres avec noms en X
plt.title("Âge des étudiants")
plt.xlabel("Nom")
plt.ylabel("Âge")
plt.show()
print("3. Courbe d'une fonction mathématique -------------------------- \n ")

# Création des valeurs X
x = np.linspace(-3, 3, 40)  # 40 points régulièrement espacés entre -3 et 3
# Calcul des valeurs Y selon la fonction y = x² - 3x + 2
y = x**2 - 3*x + 2

# Tracé de la courbe

plt.plot(x, y)
plt.title("Fonction y = x² - 3x + 2")  # Titre avec formule
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
