```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("ventes_1000.csv")

# Basic exploration
print("=== DATASET SHAPE ===")
print(f"Shape: {df.shape}")  # (rows, columns)

print("\n=== DATASET INFO ===")
df.info()

print("\n=== FIRST 5 ROWS ===")
print(df.head())

print("\n=== LAST 5 ROWS ===")
print(df.tail())

print("\n=== RANDOM 5 ROWS ===")
print(df.sample(5))

print("\n=== STATISTICAL SUMMARY ===")
print(df.describe())

print("\n=== UNIQUE VALUES PER COLUMN ===")
print(df.nunique())

print("\n=== UNIQUE VALUES PER ville ===")
print(df['Ville'].unique())

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())
# pour les statistiques descriptives
print("\n=== MEAN VALUES (NUMERIC COLUMNS ONLY) ===")
print(df.mean(numeric_only=True))

print("\n=== MEDIAN VALUES (NUMERIC COLUMNS ONLY) ===")
print(df.median(numeric_only=True))
print("\n=== STANDARD DEVIATION (NUMERIC COLUMNS ONLY) ===")
print(df.std(numeric_only=True))

print("\n=== MINIMUM VALUES ===")
print(df.min(numeric_only=True))

print("\n=== MAXIMUM VALUES ===")
print(df.max(numeric_only=True))

print("\n=== DUPLICATED ROWS ===")
duplicated_count = df.duplicated().sum()
print(duplicated_count)
df.drop_duplicates(inplace=True) #upprime les lignes en double (doublons) d'un DataFrame pandas, en modifiant directement le DataFrame original.
df.dropna()



df["Ville"] = df["Ville"].fillna(df["Ville"].mode()[0])
df["Produit"] = df["Produit"].fillna(df["Produit"].mode()[0])
df["Quantite"] = df["Quantite"].fillna(df["Quantite"].mean())
df["Prix"] = df["Prix"].fillna(df["Prix"].mean())
df["Chiffre_Affaires"] = df["Chiffre_Affaires"].fillna(df["Chiffre_Affaires"].mean())
df["ID_Client"] = df["ID_Client"].fillna(df["ID_Client"].mean())

df["Date"]=pd.to_datetime(df["Date"])

df["Quantite"] = df["Quantite"].astype(int)


Q1 = df["Chiffre_Affaires"].quantile(0.25)
Q3 = df["Chiffre_Affaires"].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df["Chiffre_Affaires"] < Q1 - 1.5*IQR) | (df["Chiffre_Affaires"] > Q3 + 1.5*IQR)]
#Normalisation
df["Quantite_norm"] = (df["Quantite"] - df["Quantite"].min()) / (df["Quantite"].max() - df["Quantite"].min())
#standarisation
df["Quantite_std"] = (df["Quantite"] - df["Quantite"].mean()) / df["Quantite"].std()
print(df)

#6. Encoder les variables cat√©gorielles

df_encoded=pd.get_dummies(df, columns=["Ville", "Produit"])
df_encoded
df_final = df_encoded.drop(columns=["ID_Client"])
df_final.to_csv("ventes_1000_nettoyes.csv",index=False)
```
