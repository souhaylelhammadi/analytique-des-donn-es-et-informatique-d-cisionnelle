```python
import pandas as pd

df=pd.read_csv('Titanic-Dataset.csv')
df.shape
df.columns
df.dtypes
df.head()
df.loc[df['PassengerId'] == 1, 'Age'] = 30
df.head()
passagers_filtres = df[
    (df['Pclass'] == 1) &
    (df['Survived'] == 1) &
    (df['Age'] > 30)
]

passagers_filtres.shape[0]
df.groupby('Pclass')['Fare'].agg(['mean', 'std', 'max'])
```
