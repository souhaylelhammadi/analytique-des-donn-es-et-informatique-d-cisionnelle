```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Chargement des dataset
df=pd.read_csv('Titanic-Dataset.csv')
# Dimensions du dataset
print("\n Dimensions du dataset :", df.shape)
# Colonnes
print("\n Colonnes :", df.columns.tolist())
# Types de données
print("\n Types de données :",df.dtypes)
# 5 premières lignes
print("\n 5 premières lignes :",df.head())
```

```python

#1. Mise à jour conditionnelle : Mettez à jour l'âge d'un passager spécifique (PassengerId= 1) à 30 ans.
df.loc[df['PassengerId']==1,'Age'] =30
print(df.loc[df['PassengerId'] == 1, ['PassengerId', 'Age']])
```

```python
#2. Filtrage avec conditions multiples : Trouvez tous les passagers de première classe qui ont survécu et ont plus de 30 ans.
filtered = df[(df['Pclass'] == 1) & (df['Survived'] == 1) & (df['Age'] > 30)]
print(f"Nombre de passagers trouvés : {len(filtered)}")
print(filtered[['PassengerId', 'Name', 'Age', 'Pclass', 'Survived']])
```

```python
#3. Agrégation de données avec groupby : Calculez la moyenne, l'écart-type et le maximum du prix des billets par classe.
agg = df.groupby('Pclass')['Fare'].agg(['mean', 'std', 'max'])
print(agg)
```

```python
#4. Gestion des valeurs manquantes : Identifiez les colonnes avec des valeurs manquantes et remplacez l'âge manquant par la médiane d'âge de chaque classe.
missing = df.isnull().sum()
print("Valeurs manquantes par colonne :")
print(missing[missing > 0])
# Remplacer l'âge manquant par la médiane d'âge par classe
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
print("\nValeurs manquantes dans 'Age' après traitement :", df['Age'].isnull().sum())

```

```python
#5. Fusion de DataFrames : Créez un DataFrame séparé avec des informations supplémentaires et fusionnez-le.
extra_info = pd.DataFrame({
    'PassengerId': [1, 2, 3],
    'Nationality': ['British', 'American', 'French']
})
# Fusion
merged = pd.merge(df, extra_info, on='PassengerId', how='left')
print(merged[['PassengerId', 'Name', 'Nationality']].head())
#On fusionne avec pd.merge() sur la clé PassengerId en gardant toutes les lignes de df.
```

```python
#6. Manipulation de dates : Bien que le Titanic n'ait pas de vraies dates, simulez une analyse temporelle.
df['SimulatedDate'] = pd.date_range(start='1912-04-10', periods=len(df), freq='D')
print(df[['PassengerId', 'SimulatedDate']].head())
#On crée une série de dates fictives pour simuler une analyse temporelle.
```

```python
#7. Application de fonctions personnalisées : Catégorisez les passagers par tranche d'âge.
def age_category(age):
    if age < 18:
        return 'Enfant'
    elif age < 60:
        return 'Adulte'
    else:
        return 'Senior'

df['AgeCategory'] = df['Age'].apply(age_category)
print(df[['Age', 'AgeCategory']].head())
```

```python
#8. Pivotage de données : Créez un tableau croisé du taux de survie par classe et sexe.
pivot = df.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean')
print(pivot)
#On crée un tableau croisé dynamique avec pivot_table() qui calcule le taux de survie moyen par classe et sexe.
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Distribution des âges
axes[0, 0].hist(df['Age'].dropna(), bins=30, edgecolor='black')
axes[0, 0].set_title('Distribution des âges')

# 2. Survie par classe
df.groupby('Pclass')['Survived'].mean().plot(kind='bar', ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Taux de survie par classe')

# 3. Prix des billets par classe
sns.boxplot(x='Pclass', y='Fare', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Prix des billets par classe')

# 4. Répartition hommes/femmes
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1, 1])
axes[1, 1].set_title('Répartition hommes/femmes')

plt.tight_layout()
plt.show()
```

```python
#10 Graphiques à barres empilées : Créez un diagramme à barres empilées montrant la survie par classe.
stacked=df.groupby(['Pclass','Survived']).size().unstack()
stacked.plot(kind='bar',stacked=True,color=['red','green'])
plt.title('survie par class (empilé)')
plt.xlabel('classe')
plt.ylabel('Nombre de passagers')
plt.legend(['Non-Survivants','Survivants'])
plt.show()

```

```python
#11 11. Annotations sur un graphique : Ajoutez des annotations sur le graphique des prix des billets.
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
max_fare = df.loc[df['Fare'].idxmax()]
plt.annotate(f'Max: {max_fare["Fare"]:.2f}',
             xy=(max_fare['Age'], max_fare['Fare']),
             xytext=(max_fare['Age']+5, max_fare['Fare']+10),
             arrowprops=dict(arrowstyle='->'))
plt.title('Prix des billets vs Âge')
plt.xlabel('Âge')
plt.ylabel('Prix du billet')
plt.show()
```

```python
#12Échelles logarithmiques : Visualisez la distribution des prix des billets avec une échelle logarithmique.
plt.figure(figsize=(8, 5))
plt.hist(df['Fare'], bins=50, edgecolor='black', log=True)
plt.title('Distribution des prix des billets (échelle logarithmique)')
plt.xlabel('Prix')
plt.ylabel('Fréquence (log)')
plt.show()
```

```python
#13  Légende avancée : Créez un scatter plot avec une légende personnalisée.
plt.figure(figsize=(10, 6))
for survived, group in df.groupby('Survived'):
    plt.scatter(group['Age'], group['Fare'],
                label='Survivant' if survived == 1 else 'Non-Survivant',
                alpha=0.6)
plt.legend(title='Survie')
plt.title('Âge vs Prix du billet')
plt.xlabel('Âge')
plt.ylabel('Prix')
plt.show()
```

```python
#Graphique en camembert : un camembert montrant la répartition des ports d'embarquement (Southampton S (Angleterre), Cherbourg C (France) et Queenstown Q (Irlande)).
df['Embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%',
                                   colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Répartition des ports d’embarquement')
plt.ylabel('')
plt.show()
```

```python
#15 Personnalisation des ticks : Personnalisez les ticks d'un histogramme des âges.
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution des âges avec ticks personnalisés')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.xticks(np.arange(0, 100, 10))
plt.yticks(np.arange(0, 100, 10))
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Age', data=df, palette='muted')
plt.title('Distribution des âges par classe (Violin Plot)')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap des corrélations')
plt.show()
```

```python
g = sns.FacetGrid(df, col='Pclass', row='Survived', height=4)
g.map(sns.scatterplot, 'Age', 'Fare', alpha=0.6)
g.add_legend()
plt.show()
```

```python
sns.jointplot(x='Age', y='Fare', data=df, kind='scatter', alpha=0.5)
plt.suptitle('Relation conjointe Âge vs Prix du billet', y=1.02)
plt.show()
```

```python
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True, common_norm=False, palette=['red', 'green'])
plt.title('Densité de l’âge par survie')
plt.xlabel('Âge')
plt.ylabel('Densité')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']],
             hue='Survived', palette={0:'red', 1:'green'})
plt.suptitle('Pairplot des variables numériques (hue=Survived)', y=1.02)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

corr = df[numeric_cols].corr()
linkage_matrix = linkage(corr, method='ward')
sns.clustermap(corr, annot=True, cmap='coolwarm', row_linkage=linkage_matrix, col_linkage=linkage_matrix)
plt.suptitle('Heatmap avec clustering hiérarchique', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df, palette='Set2')
plt.title('Distribution de l’âge par classe et sexe')
plt.show()
```
